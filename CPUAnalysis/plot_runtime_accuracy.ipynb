```python
import matplotlib.pyplot as plt
import numpy as np

file11 = "../CPUAnalysis/CPUvsAccuracy/Q1/Q1_GS_Results/cpuanalysis0.csv"
file12 = "../CPUAnalysis/CPUvsAccuracy/Q1/Q1_J_Results/cpuanalysis0.csv"
file13 = "../CPUAnalysis/CPUvsAccuracy/Q1/Q1_SOR_Results/cpuanalysis0.csv"
file21 = "../CPUAnalysis/CPUvsAccuracy/Q2/Q2_GS_Results/cpuanalysis0.csv"
file22 = "../CPUAnalysis/CPUvsAccuracy/Q2/Q2_J_Results/cpuanalysis0.csv"
file23 = "../CPUAnalysis/CPUvsAccuracy/Q2/Q2_SOR_Results/cpuanalysis0.csv"
file31 = "../CPUAnalysis/CPUvsAccuracy/Q3/Q3_GS_Results/cpuanalysis0.csv"
file32 = "../CPUAnalysis/CPUvsAccuracy/Q3/Q3_J_Results/cpuanalysis0.csv"
file33 = "../CPUAnalysis/CPUvsAccuracy/Q3/Q3_SOR_Results/cpuanalysis0.csv"

files = [[file11, file12, file13],
         [file21, file22, file23],
         [file31, file32, file33]]


for fileset in files:
    GS = np.loadtxt(fileset[0],delimiter=", ")
    J = np.loadtxt(fileset[1],delimiter=", ")
    SOR = np.loadtxt(fileset[2],delimiter=", ")

    plt.plot(GS[:,0], GS[:,3],label="GS")
    plt.plot(J[:,0], J[:,3],label="J")
    plt.plot(SOR[:,0], SOR[:,3],label="SOR")
    plt.legend()
    
    plt.xlabel("Accuracy Threshold")
    plt.ylabel("CPU Active Time (s/100)")
    plt.xscale("log")
    plt.yscale("log")

    #plt.savefig('CPUcon_vs_iterations.png')
    plt.show()
```
