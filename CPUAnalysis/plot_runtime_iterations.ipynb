```python
import matplotlib.pyplot as plt
fig,a =  plt.subplots(1,2,figsize=(10,5), squeeze=False)
import numpy as np

file11 = "../CPUAnalysis/CPUvsIterations/Q1/Q1_GS_Results/cpuanalysis0.csv"
file12 = "../CPUAnalysis/CPUvsIterations/Q1/Q1_J_Results/cpuanalysis0.csv"
file13 = "../CPUAnalysis/CPUvsIterations/Q1/Q1_SOR_Results/cpuanalysis0.csv"
file21 = "../CPUAnalysis/CPUvsIterations/Q2/Q2_GS_Results/cpuanalysis0.csv"
file22 = "../CPUAnalysis/CPUvsIterations/Q2/Q2_J_Results/cpuanalysis0.csv"
file23 = "../CPUAnalysis/CPUvsIterations/Q2/Q2_SOR_Results/cpuanalysis0.csv"
file31 = "../CPUAnalysis/CPUvsIterations/Q3/Q3_GS_Results/cpuanalysis0.csv"
file32 = "../CPUAnalysis/CPUvsIterations/Q3/Q3_J_Results/cpuanalysis0.csv"
file33 = "../CPUAnalysis/CPUvsIterations/Q3/Q3_SOR_Results/cpuanalysis0.csv"

files = [[file11, file12, file13],
         [file21, file22, file23],
         [file31, file32, file33]]
i=0

for fileset in files:
    
    GS = np.loadtxt(fileset[0],delimiter=", ")
    J = np.loadtxt(fileset[1],delimiter=", ")
    SOR = np.loadtxt(fileset[2],delimiter=", ")

    a[0][0].plot(GS[:,0], GS[:,3],label="GS")
    a[0][0].plot(J[:,0], J[:,3],label="J")
    a[0][0].plot(SOR[:,0], SOR[:,3],label="SOR")
    a[0][0].legend()
    a[0][0].set_xlabel("No. of iterations")
    a[0][0].set_ylabel("Total Active CPU time (s/100)")
    a[0][0].set_xscale("log")
    a[0][0].set_yscale("log")
    
    a[0][1].plot(GS[:,0], GS[:,1],label="GS")
    a[0][1].plot(J[:,0], J[:,1],label="J")
    a[0][1].plot(SOR[:,0], SOR[:,1],label="SOR")
    a[0][1].legend()
    a[0][1].set_xlabel("No. of iterations")
    a[0][1].set_ylabel("CPU Consumption in %")
    a[0][1].set_xscale("log")
    
    plt.show()
    
    if i == 2:
        break
```
