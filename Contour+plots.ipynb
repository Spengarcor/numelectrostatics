```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname="Q1.csv"
Data = np.loadtxt(fname, delimiter=';')

X = [[i for i in range(len(Data[0,:]))] for j in range(len(Data[:,0]))]

Y = [[j for i in range(len(Data[0,:]))] for j in range(len(Data[:,0]))]

plt.contour(X, Y, Data, 15)
plt.colorbar();
plt.show();
```

```python
fname="Q2.csv"
Data2 = np.loadtxt(fname, delimiter=';')

X2 = [[i for i in range(len(Data2[0,:]))] for j in range(len(Data2[:,0]))]

Y2 = [[j for i in range(len(Data2[0,:]))] for j in range(len(Data2[:,0]))]

plt.contour(X2, Y2, Data2, 25)
plt.colorbar();
plt.show();
```
