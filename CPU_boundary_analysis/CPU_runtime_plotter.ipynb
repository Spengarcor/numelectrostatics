```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
normal = "./Q1_SOR/cpuanalysis0.csv"
modified = "./Q1_0/cpuanalysis0.csv"
hybrid = "./Q1_4/cpuanalysis0.csv"

N = np.loadtxt(normal,delimiter=", ")
M = np.loadtxt(modified,delimiter=", ")
H = np.loadtxt(hybrid,delimiter=", ")
```

```python
plt.plot(N[:,0], N[:,3],label="Normal")
plt.plot(M[:,0], M[:,3],label="Modified")
plt.plot(H[:,0], H[:,3],label="Hybrid")
plt.legend()

plt.title("CPU runtime")
plt.xlabel("number of iterations")
plt.ylabel("time (s/100)")
plt.xscale("log")
#plt.yscale("log")

plt.show()
```
